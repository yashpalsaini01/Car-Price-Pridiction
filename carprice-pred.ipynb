#cleaning

```python
df=df[df['year'].str.isnumeric()] #it remove all row having non numeric value
```

```python
df['year']=df['year'].astype(int)
```

```python
df.info()
```

```python
df[df['Price']!='Ask For Price']
```

```python
df['Price']=df['Price'].str.replace(',','')
# Use pd.to_numeric with errors='coerce' to convert non-numeric values to NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
# Optional: fill NaN values with 0 or drop rows with NaN prices
# df['Price'] = df['Price'].fillna(0)  # Replace NaN with 0
# df = df.dropna(subset=['Price'])
```

```python
df.info()
```

```python
df['kms_driven']=df['kms_driven'].str.split(' ').str.get(0)
```

```python
# First, ensure the column is treated as string type
df['kms_driven'] = df['kms_driven'].astype(str)

# Remove commas from the string values
df['kms_driven'] = df['kms_driven'].str.replace(',','')

# Filter rows where kms_driven is numeric, handling NaN values properly
# Also filter out 'nan' strings that result from NaN values being converted to string
df = df[(df['kms_driven'].str.isnumeric()) & (df['kms_driven'] != 'nan')]

# Convert to numeric type
df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors='coerce')
```

```python
df=df[~df['fuel_type'].isna()]
```

```python
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df=df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df[df['Price']>6e6] #this is a outlier
```

```python
df=df[df['Price']<6e6].reset_index(drop=True)
```

#model

```python
X=df.drop(columns='Price')
```

```python
y=df['Price']
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
oe=OneHotEncoder()
oe.fit(X[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=oe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
print(pipe)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('car.pkl','wb'))
```
